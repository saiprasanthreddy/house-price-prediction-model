```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')


pd.pandas.set_option('display.max_columns',None)
```

```python
data_set = pd.read_csv('train.csv')
data_set.head()
```

```python
data_set.shape
```

```python
feature_with_na = [feature for feature in data_set.columns if data_set[feature].isnull().sum()>1 and data_set[feature].dtypes ]
for feature in feature_with_na:
    print(feature,np.round(data_set[feature].isnull().mean(),4),'% missing values')
```

```python
for feature in feature_with_na:
    data = data_set.copy()
    data[feature] = np.where(data[feature].isnull(),1,0)


    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()




```

```python
numerical_feature = [feature for feature in data_set.columns if data_set[feature].dtypes != 'O']
print('Number of numerical variables: ', len(numerical_feature))
data_set[numerical_feature].head()
```

```python
year_feature = [feature  for feature in numerical_feature if 'Yr' in feature or 'Year' in feature]
year_feature
```

```python
for feature in year_feature:
    print(feature,data_set[feature].unique())
```

```python
data_set.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title('House Price vs Year Sold')
```

```python
for feature in year_feature:
    if feature != 'YrSold':
        data[feature] = data_set['YrSold'] - data_set[feature]

        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('Sale Price')
        plt.show()
```

```python
discrite_feature = [feature for feature in numerical_feature if len(data_set[feature].unique())<25 and feature not in year_feature+['Id']]
print('Number of discrite variables: ',len(discrite_feature))
discrite_feature
```

```python
for feature in discrite_feature:
    data_set.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('Sale Price')
    plt.title(feature)
    plt.show()
```

```python
countiues_feature = [feature for feature in numerical_feature if feature not in discrite_feature+year_feature+['Id']]
print('Number of countiues variables: ',len(countiues_feature))
countiues_feature
```

```python
for feature in countiues_feature:
    data_set[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```

```python
for feature in countiues_feature:
    data = data_set.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data['SalePrice'] = np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('Sale Price')
        plt.title(feature)
        plt.show()
```

outliers


```python
for feature in countiues_feature:
  data = data_set
  if 0 in data[feature].unique():
    pass
  else:
    data[feature] = np.log(data[feature])
    data.boxplot(column=feature)
    plt.ylabel(feature)
    plt.title(feature)
    plt.show()
```

```python
categorical_feature = [feature for feature in data_set.columns if data_set[feature].dtypes == 'O']
print('Number of categorical variables: ',len(categorical_feature))
categorical_feature
```

```python
data_set[categorical_feature].head()
```

```python
for feature in categorical_feature:
    print('The feature is {} and number of categories are {}'.format(feature,len(data_set[feature].unique())))
```

```python
for feature in categorical_feature:
    data = data_set.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('Sale Price')
    plt.title(feature)
    plt.show()
```
